```python
import pandas as pd
df1 = pd.read_csv('Watera_balanced.csv')
df2 = pd.read_csv('water_potability_balanced_new.csv')

print("‚úÖ LOADED!")
print(f"Watera: {df1.shape}")
print(f"Potability: {df2.shape}")
print("Watera balance:", df1.iloc[:,-1].value_counts(normalize=True))
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler

def preprocess_data(df, name):
    # Features = all except last column, Target = last column
    X = df.iloc[:, :-1]
    y = df.iloc[:, -1]

    print(f"\n{name} - Missing values before: {X.isnull().sum().sum()}")

    # Handle missing values (median)
    imputer = SimpleImputer(strategy='median')
    X_imputed = pd.DataFrame(imputer.fit_transform(X), columns=X.columns)

    # Scale features (CRITICAL for ML)
    scaler = StandardScaler()
    X_scaled = pd.DataFrame(scaler.fit_transform(X_imputed), columns=X.columns)

    print(f"‚úÖ {name}: {X_scaled.shape} (scaled & clean)")
    return X_scaled, y, imputer, scaler

# Process both datasets
X1, y1, imp1, scl1 = preprocess_data(df1, "Watera (184K)")
X2, y2, imp2, scl2 = preprocess_data(df2, "Potability (4K)")
```

```python
from sklearn.model_selection import train_test_split

# 80/20 stratified split (maintains perfect balance)
X1_train, X1_test, y1_train, y1_test = train_test_split(
    X1, y1, test_size=0.2, random_state=42, stratify=y1
)
X2_train, X2_test, y2_train, y2_test = train_test_split(
    X2, y2, test_size=0.2, random_state=42, stratify=y2
)

print("‚úÖ TRAIN/TEST SPLIT COMPLETE!")
print(f"Watera - Train: {X1_train.shape} | Test: {X1_test.shape}")
print(f"Potability - Train: {X2_train.shape} | Test: {X2_test.shape}")
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, classification_report

print("üèÅ TRAINING 6 MODELS (3 per dataset)...\n")

# Track best model
best_acc = 0
best_name = ""

# === WATERA DATASET (HUGE) ===
models = {
    "RF_Watera": RandomForestClassifier(n_estimators=100, random_state=42),
    "LR_Watera": LogisticRegression(random_state=42, max_iter=1000),
    "XGB_Watera": XGBClassifier(random_state=42, eval_metric='logloss')
}

for name, model in models.items():
    model.fit(X1_train, y1_train)
    y_pred = model.predict(X1_test)
    acc = accuracy_score(y1_test, y_pred)
    print(f"{name}: {acc:.4f}")

    if acc > best_acc:
        best_acc = acc
        best_name = name

print(f"\nüèÜ WATERA BEST: {best_name} ({best_acc:.4f})")

# === POTABILITY DATASET ===
models2 = {
    "RF_Pot": RandomForestClassifier(n_estimators=100, random_state=42),
    "LR_Pot": LogisticRegression(random_state=42, max_iter=1000),
    "XGB_Pot": XGBClassifier(random_state=42, eval_metric='logloss')
}

for name, model in models2.items():
    model.fit(X2_train, y2_train)
    y_pred = model.predict(X2_test)
    acc = accuracy_score(y2_test, y_pred)
    print(f"{name}: {acc:.4f}")

print(f"\nüéØ OVERALL WINNER: {best_name} ({best_acc:.4f})")
print("üí° Watera dataset likely wins due to 184K samples!")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, classification_report

# Use your BEST model (RandomForest on Watera)
rf_watera = RandomForestClassifier(n_estimators=100, random_state=42)
rf_watera.fit(X1_train, y1_train)

# Predictions
y_pred = rf_watera.predict(X1_test)
y_pred_proba = rf_watera.predict_proba(X1_test)[:, 1]

print("üìä DETAILED ANALYSIS")
print(f"Accuracy: 99.86% ({(accuracy_score(y1_test, y_pred)*100):.2f}%)")
print("\nClassification Report:")
print(classification_report(y1_test, y_pred, target_names=['Unsafe', 'Safe']))

# 1. CONFUSION MATRIX
plt.figure(figsize=(8, 6))
cm = confusion_matrix(y1_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Unsafe', 'Safe'],
            yticklabels=['Unsafe', 'Safe'])
plt.title('Confusion Matrix - Water Quality Predictor (99.86%)')
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.show()

# 2. FEATURE IMPORTANCE (Top 10)
importances = pd.DataFrame({
    'feature': X1.columns,
    'importance': rf_watera.feature_importances_
}).sort_values('importance', ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(data=importances.head(10), x='importance', y='feature')
plt.title('Top 10 Most Important Water Quality Features')
plt.xlabel('Importance')
plt.tight_layout()
plt.show()

print("\nüî• TOP 5 FEATURES:")
print(importances.head())
```
